```python
site_map = folium.Map(location=[spacex_df['Lat'].mean(), spacex_df['Long'].mean()], zoom_start=5)

# Create a marker cluster to group nearby points (for easier viewing)
marker_cluster = MarkerCluster().add_to(site_map)

# Loop through each launch record
for index, record in spacex_df.iterrows():
    lat = record['Lat']
    lon = record['Long']
    launch_site = record['Launch Site']
    outcome = record['class']
    
    # Set color based on success/failure
    color = 'green' if outcome == 1 else 'red'
    
    # Create marker with popup info
    folium.CircleMarker(
        location=[lat, lon],
        radius=5,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.7,
        popup=f"{launch_site}, Outcome: {'Success' if outcome == 1 else 'Failure'}"
    ).add_to(marker_cluster)

# Display the map
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import sin, cos, sqrt, atan2, radians

# Function to calculate distance between two coordinates
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # approximate radius of Earth in km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1)*cos(lat2)*sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    return R * c
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Coordinates for the launch site (example: CCAFS LC-40)
launch_site_lat = 28.562302
launch_site_lon = -80.577356

# Coordinates for the closest coastline (point at the shore)
coastline_lat = 28.56320
coastline_lon = -80.56700
```

```python
# TASK: Explore the generated folium map and show the proximities

import folium
from folium.features import DivIcon
from math import sin, cos, sqrt, atan2, radians

# Define distance calculation function
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6373.0  # radius of Earth in km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    return R * c

# Coordinates for CCAFS LC-40 Launch Site
launch_site_lat = 28.562302
launch_site_lon = -80.577356

# Coordinates for nearby features
coastline_lat, coastline_lon = 28.56320, -80.56700     # nearest coastline
railway_lat, railway_lon = 28.57210, -80.58560         # nearest railway line
highway_lat, highway_lon = 28.56300, -80.57000         # nearest highway

# Calculate distances (in km)
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
distance_railway = calculate_distance(launch_site_lat, launch_site_lon, railway_lat, railway_lon)
distance_highway = calculate_distance(launch_site_lat, launch_site_lon, highway_lat, highway_lon)

# Create a folium map centered at the launch site
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=14)

# Launch site marker
folium.CircleMarker(
    [launch_site_lat, launch_site_lon],
    radius=7,
    color='red',
    fill=True,
    fill_color='red',
    popup='Launch Site: CCAFS LC-40'
).add_to(site_map)

# Coastline marker + line
folium.Marker(
    [coastline_lat, coastline_lon],
    icon=DivIcon(html=f'<div style="font-size: 12px; color:blue;"><b>{distance_coastline:.2f} KM</b></div>')
).add_to(site_map)
folium.PolyLine([[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]], color='blue', weight=2).add_to(site_map)

# Railway marker + line
folium.Marker(
    [railway_lat, railway_lon],
    icon=DivIcon(html=f'<div style="font-size: 12px; color:green;"><b>{distance_railway:.2f} KM</b></div>')
).add_to(site_map)
folium.PolyLine([[launch_site_lat, launch_site_lon], [railway_lat, railway_lon]], color='green', weight=2).add_to(site_map)

# Highway marker + line
folium.Marker(
    [highway_lat, highway_lon],
    icon=DivIcon(html=f'<div style="font-size: 12px; color:orange;"><b>{distance_highway:.2f} KM</b></div>')
).add_to(site_map)
folium.PolyLine([[launch_site_lat, launch_site_lon], [highway_lat, highway_lon]], color='orange', weight=2).add_to(site_map)

# Display the map
site_map
```
